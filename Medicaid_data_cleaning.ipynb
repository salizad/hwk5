```r
# Read data
kff.dat <- read_excel("/home/salizad/econ470/a0/work/hwk5/reformatted_kff_data.xlsx")

# Manually create the expanded variable based on your known states
kff.final <- kff.dat %>%
  mutate(expanded = case_when(
    State %in% c("Alaska", "Arizona", "Arkansas", "California", "Colorado", 
                  "Connecticut", "Delaware", "District of Columbia", "Hawaii", 
                  "Idaho", "Illinois", "Indiana", "Iowa", "Kentucky", 
                  "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", 
                  "Minnesota", "Missouri", "Montana", "Nebraska", "Nevada", 
                  "New Hampshire", "New Jersey", "New Mexico", "New York", 
                  "North Carolina", "North Dakota", "Ohio", "Oklahoma", 
                  "Oregon", "Pennsylvania", "Rhode Island", "South Dakota", 
                  "Utah", "Vermont", "Virginia", "Washington", "West Virginia") ~ "Adopted",
    TRUE ~ "Not Adopted"
  )) %>%
  mutate(date_adopted = na_if(date_adopted, "N/A")) %>%
  mutate(date_adopted = mdy(date_adopted)) %>%
  select(State, expanded, date_adopted)
 
# Display the cleaned data
print(kff.final)
```

```r
# Read data from Excel
kff.dat <- read_excel("/home/salizad/econ470/a0/work/hwk5/reformatted_kff_data.xlsx")
 
# Clean data and create the expanded variable
kff.final <- kff.dat %>%
  mutate(
    expanded = ifelse(date_adopted != "N/A" & !is.na(date_adopted), "Adopted", "Not Adopted"),  # Set expanded based on date_adopted
    date_adopted = na_if(date_adopted, "N/A")  # Handle "N/A" entries
  ) %>%
  mutate(date_adopted = mdy(date_adopted)) %>%  # Convert to date format
  select(State, expanded, date_adopted)  # Select relevant columns
 
# Display the cleaned data
print(kff.final)
```

```r
# Read data from Excel
kff.dat <- read_excel("/home/salizad/econ470/a0/work/hwk5/reformatted_kff_data.xlsx")
 
# Inspect unique values in date_adopted
unique_dates <- unique(kff.dat$date_adopted)
print(unique_dates)
 
# Clean data and create the expanded variable
kff.final <- kff.dat %>%
  mutate(
    # Handle "N/A" entries by converting them to NA
    date_adopted = na_if(date_adopted, "N/A"),
    # Trim whitespace from the date strings
    date_adopted = trimws(date_adopted),
    # Convert to Date format: handle numeric entries separately
    date_adopted = case_when(
      grepl("^[0-9]+$", date_adopted) ~ as.Date(as.numeric(date_adopted), origin = "1970-01-01"),  # Numeric values
      TRUE ~ as.Date(date_adopted, format = "%Y-%m-%d")  # Existing date format
    )
  ) %>%
  # Filter out rows with NA in date_adopted
  filter(!is.na(date_adopted)) %>%
  # Create the expanded variable based on the cleaned date_adopted
  mutate(
    expanded = ifelse(date_adopted < Sys.Date(), "Adopted", "Not Adopted")
  ) %>%
  # Select relevant columns
  select(State, expanded, date_adopted)
 
# Display the cleaned data
print(kff.final)
```
