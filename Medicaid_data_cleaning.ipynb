```r
options(repos = c(CRAN = "https://cloud.r-project.org"))

if (!require("pacman")) install.packages("pacman")
pacman::p_load(tidyverse, ggplot2, dplyr, lubridate)
library(readr)
library(magrittr)
library(stringr)
library(tidyr)
```

```r
# Read the CSV file, skipping the first two rows
df <- read.csv("/home/salizad/econ470/a0/work/raw_data.csv", skip = 2, stringsAsFactors = FALSE)
# Remove rows that contain summary information
df <- df[-1, ]  # Remove the first row (summary for "United States")
 
# Remove rows with all NA values (if any)
df <- df[rowSums(is.na(df)) != ncol(df), ]
 
# Clean the data by selecting relevant columns and renaming them
df_cleaned <- df %>%
  select(state = Location,
         expanded = Status.of.Medicaid.Expansion.Decision,
         date_adopted = Expansion.Implementation.Date) %>%
  mutate(expanded = ifelse(grepl("Adopted", expanded), "TRUE", "FALSE"),
         date_adopted = ifelse(date_adopted == "" | date_adopted == "N/A", "NA", date_adopted))
 
# Save as a TXT file
write.table(df_cleaned, file = "medicaid_data.txt", sep = "\t", row.names = FALSE, quote = FALSE)
print(head(df))
```
