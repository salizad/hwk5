```r
options(repos = c(CRAN = "https://cloud.r-project.org"))

if (!require("pacman")) install.packages("pacman")
pacman::p_load(tidyverse, ggplot2, dplyr, lubridate)
library(readr)
library(magrittr)
library(stringr)
library(tidyr)

 
# Load the data
hcris_data <- read_csv('../data/output/HCRIS_Data_v2010.csv')
# Read the KFF data; adjust if the first row isn't the headers
kff_data <- read_csv("/home/salizad/econ470/a0/work/raw_data.csv", skip = 3)
```

# Merge the Data 

```r

colnames(kff_data) <- c("state", "expansion_status", "expansion_date", "medicaid_expansion", "other_info")
 kff_data <- kff_data %>%
  rename_with(~ trimws(.))
head(kff_data)
```

```r
# Simplify the expansion status to a binary format (Adopted or Not Adopted)
kff_data <- kff_data %>%
  mutate(expansion_binary = ifelse(expansion_status == "Adopted", 1, 0))
 
# Select only the columns you need
kff_cleaned <- kff_data %>%
  select(state, expansion_binary, expansion_date)
 
# Check the cleaned and simplified KFF data
head(kff_cleaned)
```

```r
# Assuming hcris_data is already loaded and cleaned
merged_data <- hcris_data %>%
  left_join(kff_cleaned, by = "state")
 
# Check the merged data
head(merged_data)
```

```r
# Clean state names in both datasets
hcris_data <- hcris_data %>%
  mutate(state = str_trim(state))
 
kff_cleaned <- kff_cleaned %>%
  mutate(state = str_trim(state))
 
# Reattempt the merge
merged_data <- hcris_data %>%
  left_join(kff_cleaned, by = "state")
```

```r
# Summarize mean uncompensated care by year
mean_uncompensated_care <- merged_data %>%
  group_by(year) %>%
  summarise(mean_care = mean(tot_uncomp_care_charges, na.rm = TRUE) / 1e6)  # Use the correct column name
 
# View the result
print(mean_uncompensated_care)
```

```r
# Summarize mean uncompensated care and partial payments by year
mean_uncompensated_care <- merged_data %>%
  group_by(year) %>%
  summarise(mean_care = mean(tot_uncomp_care_charges + tot_uncomp_care_partial_pmts, na.rm = TRUE) / 1e6)
 
# View the result
print(mean_uncompensated_care)
```

```r
# Define the list of states that expanded Medicaid in 2014
expanded_states_list <- c("CA", "NY", "KY")  # Replace with actual state abbreviations
 
# Filter data for expanded and never expanded states
expanded_states <- merged_data %>%
  filter(state %in% expanded_states_list & year <= 2014)
 
never_expanded_states <- merged_data %>%
  filter(!state %in% expanded_states_list)
 
# Combine datasets for easier plotting
filtered_data <- bind_rows(
  expanded_states %>% mutate(expansion_status = "Expanded"),
  never_expanded_states %>% mutate(expansion_status = "Never Expanded")
)
 
# Calculate mean uncompensated care by year and expansion status
mean_uncomp_care <- filtered_data %>%
  group_by(year, expansion_status) %>%
  summarize(mean_uncomp_care = mean(tot_uncomp_care_charges, na.rm = TRUE))
 
# Plotting
ggplot(mean_uncomp_care, aes(x = year, y = mean_uncomp_care, color = expansion_status)) +
  geom_line() +
  geom_point() +
  labs(title = "Mean Uncompensated Care Over Time",
       x = "Year",
       y = "Mean Uncompensated Care Charges") +
  theme_minimal() +
  scale_color_manual(values = c("blue", "red")) +
  theme(legend.title = element_blank())
```

The graph illustrates mean uncompensated care charges over time for states that expanded Medicaid in 2014 (blue line) versus those that never expanded (red line). Before the expansion, uncompensated care was relatively high and stable for both groups, with expanded states showing slightly lower charges. Post-2014, expanded states experienced a significant drop in uncompensated care, suggesting that Medicaid expansion positively impacted financial burdens. In contrast, never-expanded states saw an increase in uncompensated care, indicating worsening conditions. This divergence suggests a potential treatment effect, where Medicaid expansion leads to reduced uncompensated care charges compared to states that did not expand.
